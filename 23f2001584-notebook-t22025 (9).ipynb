```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, KFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error, roc_auc_score
from xgboost import XGBClassifier, XGBRegressor
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.feature_selection import SelectFromModel
import re


# Set random seed for reproducibility
np.random.seed(42)

# Load the datasets
try:
    train_df = pd.read_csv('/kaggle/input/engage-2-value-from-clicks-to-conversions/train_data.csv')
    test_df = pd.read_csv('/kaggle/input/engage-2-value-from-clicks-to-conversions/test_data.csv')
    sample_submission = pd.read_csv('/kaggle/input/engage-2-value-from-clicks-to-conversions/sample_submission.csv')
except FileNotFoundError:
    print("Ensure data files are in the specified path or adjust the paths.")
    # Fallback for local testing if files are not in Kaggle directory structure
    try:
        train_df = pd.read_csv('./train_data.csv')
        test_df = pd.read_csv('./test_data.csv')
        sample_submission = pd.read_csv('./sample_submission.csv')
    except FileNotFoundError:
        print("Data files not found in current directory either. Please check paths.")
        exit()

print("Train data shape (before cleaning):", train_df.shape)
print("Test data shape (before cleaning):", test_df.shape)

def clean_col_names(df):
    """Clean column names by removing special characters and standardizing."""
    cols = df.columns
    new_cols = []
    for col in cols:
        new_col = re.sub(r'[^\w\s\.]', '', col)
        new_col = re.sub(r'[\s\._]+', '_', new_col).strip('_').lower()
        new_cols.append(new_col)
    df.columns = new_cols
    return df

train_df = clean_col_names(train_df.copy())
test_df = clean_col_names(test_df.copy())

print("\nTrain data columns after cleaning:", train_df.columns.tolist())
print("Test data columns after cleaning:", test_df.columns.tolist())

for df in [train_df, test_df]:
    df.replace('not available in demo dataset', np.nan, inplace=True)
    df.replace('(not set)', np.nan, inplace=True)
    df.replace('(not provided)', np.nan, inplace=True)

numeric_cols_to_convert = [
    'purchasevalue',
    'totals_bounces',
    'totalhits',
    'pageviews',
    'totals_visits',
    'sessionnumber',
    'screensize',
    'browserm_major'
]

for df in [train_df, test_df]:
    for col in numeric_cols_to_convert:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

for df in [train_df, test_df]:
    if 'trafficsource_istruedirect' in df.columns:
        df['trafficsource_istruedirect'] = df['trafficsource_istruedirect'].astype(str).str.lower().map({'true': True, 'false': False, 'nan': False}).fillna(False).astype(int)

def create_features(df):
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], format='%Y%m%d', errors='coerce')
        df['month'] = df['date'].dt.month
        df['day'] = df['date'].dt.day
        df['weekday'] = df['date'].dt.weekday
        df['weekofyear'] = df['date'].dt.isocalendar().week.astype('Int64')
    else:
        df['month'] = np.nan
        df['day'] = np.nan
        df['weekday'] = np.nan
        df['weekofyear'] = np.nan

    if 'totalhits' in df.columns and 'pageviews' in df.columns:
        df['hits_per_pageview'] = df['totalhits'] / (df['pageviews'] + 1)
    else:
        df['hits_per_pageview'] = np.nan

    if 'totals_bounces' in df.columns and 'totals_visits' in df.columns:
        df['bounces_per_visit'] = df['totals_bounces'] / (df['totals_visits'] + 1)
    else:
        df['bounces_per_visit'] = np.nan
    return df

train_df = create_features(train_df.copy())
test_df = create_features(test_df.copy())

TARGET = 'purchasevalue'
TRAIN_CLASSIFY_TARGET = 'is_purchase'

if TARGET in train_df.columns:
    train_df[TRAIN_CLASSIFY_TARGET] = (train_df[TARGET] > 0).astype(int)
else:
    print(f"Warning: '{TARGET}' column not found in training data. Exiting.")
    exit()

drop_cols = ['date', 'userid', 'sessionid', 'gclidpresent']

train_cols_to_drop = [col for col in drop_cols if col in train_df.columns]
test_cols_to_drop = [col for col in drop_cols if col in test_df.columns]

train_df.drop(columns=train_cols_to_drop, inplace=True)
test_df.drop(columns=test_cols_to_drop, inplace=True)

X_train_full = train_df.drop(columns=[TARGET, TRAIN_CLASSIFY_TARGET])
# Corrected typo here
y_train_classify_full = train_df[TRAIN_CLASSIFY_TARGET]
y_train_regress_full = train_df[TARGET]

X_test_final_submission = test_df.copy()

X_train_cols = set(X_train_full.columns)
X_test_cols = set(X_test_final_submission.columns)
common_cols = list(X_train_cols.intersection(X_test_cols))

X_train_full = X_train_full[common_cols]
X_test_final_submission = X_test_final_submission[common_cols]

numerical_features = X_train_full.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_train_full.select_dtypes(include='object').columns.tolist()

print("\nNumerical Features (after final selection):", numerical_features)
print("Categorical Features (after final selection):", categorical_features)

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)

print("\nPreprocessing setup complete.")

# --- Train-Validation Split for Model Comparison ---
X_train, X_val, y_train_classify, y_val_classify, y_train_regress, y_val_regress = \
    train_test_split(X_train_full, y_train_classify_full, y_train_regress_full,
                     test_size=0.2, stratify=y_train_classify_full, random_state=42)

print(f"\nTrain set size: {len(X_train)} | Validation set size: {len(X_val)}")
print(f"Purchases in train_clf: {y_train_classify.sum()} ({y_train_classify.mean()*100:.2f}%)")
print(f"Purchases in val_clf: {y_val_classify.sum()} ({y_val_classify.mean()*100:.2f}%)")

# --- Define Models and Hyperparameters for Comparison (XGBoost + 2 Scikit-learn) ---

# Classifiers
classifiers = {
    'XGBClassifier': {
        'model': XGBClassifier(objective='binary:logistic', eval_metric='logloss', use_label_encoder=False, random_state=42, n_jobs=-1),
        'params': {
            'selectfrommodel__threshold': ['mean', 'median', -np.inf],
            'classifier__n_estimators': [300, 500], # Reduced for quicker search
            'classifier__learning_rate': [0.01, 0.05], # Reduced for quicker search
            'classifier__max_depth': [5, 7], # Reduced for quicker search
            'classifier__subsample': [0.8, 1.0],
            'classifier__colsample_bytree': [0.8, 1.0],
            'classifier__reg_alpha': [0, 0.1],
            'classifier__reg_lambda': [0, 0.5],
            'classifier__min_child_weight': [1, 5],
            'classifier__gamma': [0, 0.1],
            'classifier__scale_pos_weight': [1, (y_train_classify.value_counts()[0]/y_train_classify.value_counts()[1])]
        }
    },
    'RandomForestClassifier': {
        'model': RandomForestClassifier(random_state=42, n_jobs=-1, class_weight='balanced'),
        'params': {
            'selectfrommodel__threshold': ['mean', 'median', -np.inf], # SelectFromModel can be less useful with RF if not based on feature importance of itself
            'classifier__n_estimators': [100, 200], # Reduced for quicker search
            'classifier__max_depth': [10, 20, None], # Reduced for quicker search
            'classifier__min_samples_split': [2, 5],
            'classifier__min_samples_leaf': [1, 2]
        }
    },
    'GradientBoostingClassifier': {
        'model': GradientBoostingClassifier(random_state=42),
        'params': {
            'selectfrommodel__threshold': ['mean', 'median', -np.inf],
            'classifier__n_estimators': [100, 200], # Reduced for quicker search
            'classifier__learning_rate': [0.01, 0.05], # Reduced for quicker search
            'classifier__max_depth': [3, 5], # Reduced for quicker search
            'classifier__subsample': [0.8, 1.0],
            'classifier__max_features': ['sqrt', None]
        }
    }
}

# Regressors (trained on log1p transformed purchase values > 0)
regressors = {
    'XGBRegressor': {
        'model': XGBRegressor(objective='reg:squarederror', eval_metric='rmse', use_label_encoder=False, random_state=42, n_jobs=-1),
        'params': {
            'selectfrommodel__threshold': [0.005, 0.01],  # Less aggressive feature selection
            'regressor__n_estimators': [1200, 1500, 2000],  # More trees
            'regressor__learning_rate': [0.005, 0.01, 0.03],  # Smaller learning rates
            'regressor__max_depth': [12, 15, 18, 20],  # Deeper trees
            'regressor__subsample': [0.7, 0.8, 0.9],  # More randomness
            'regressor__colsample_bytree': [0.7, 0.8, 0.9],  # More column sampling
            'regressor__reg_alpha': [0, 0.0001, 0.001],  # Very light L1 regularization
            'regressor__reg_lambda': [0, 0.01, 0.1],  # Very light L2 regularization
            'regressor__min_child_weight': [1],  # Less restrictive
            'regressor__gamma': [0, 0.01],  # Minimal gamma
            'regressor__grow_policy': ['depthwise'],  # Favor deeper trees
            'regressor__max_bin': [512, 1024],  # More bins for histogram
            'regressor__tree_method': ['hist']
        }
    },
    'RandomForestRegressor': {
        'model': RandomForestRegressor(random_state=42, n_jobs=-1),
        'params': {
            'selectfrommodel__threshold': [0.005, 0.01],
            'regressor__n_estimators': [100, 200], # Reduced for quicker search
            'regressor__max_depth': [10, 20, None], # Reduced for quicker search
            'regressor__min_samples_split': [2, 5],
            'regressor__min_samples_leaf': [1, 2]
        }
    },
    'GradientBoostingRegressor': {
        'model': GradientBoostingRegressor(random_state=42),
        'params': {
            'selectfrommodel__threshold': [0.005, 0.01],
            'regressor__n_estimators': [100, 200], # Reduced for quicker search
            'regressor__learning_rate': [0.01, 0.05], # Reduced for quicker search
            'regressor__max_depth': [3, 5], # Reduced for quicker search
            'regressor__subsample': [0.8, 1.0],
            'regressor__max_features': ['sqrt', None]
        }
    }
}

best_clf_model = None
best_clf_score = -np.inf
best_clf_name = ""
clf_results = {}

print("\n--- Comparing Classification Models ---")
for name, config in classifiers.items():
    print(f"\nTraining {name} Classifier...")
    
    # Conditional pipeline creation for SelectFromModel
    if name == 'XGBClassifier':
        # SelectFromModel with XGBoost for feature importance
        clf_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('selectfrommodel', SelectFromModel(estimator=XGBClassifier(random_state=42, n_jobs=-1, use_label_encoder=False))),
            ('classifier', config['model'])
        ])
    elif name in ['RandomForestClassifier', 'GradientBoostingClassifier']:
        # SelectFromModel with RandomForest for feature importance for other tree-based models
        clf_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('selectfrommodel', SelectFromModel(estimator=RandomForestClassifier(random_state=42, n_jobs=-1))),
            ('classifier', config['model'])
        ])
    else: # Fallback if other non-tree models were to be added
        clf_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('classifier', config['model'])
        ])

    clf_search = RandomizedSearchCV(
        clf_pipeline,
        config['params'], # Now 'config['params']' will contain appropriate 'selectfrommodel__threshold'
        n_iter=15,
        cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=42),
        scoring='roc_auc',
        verbose=1,
        n_jobs=-1,
        random_state=42
    )
    clf_search.fit(X_train, y_train_classify)

    current_score = clf_search.best_score_
    clf_results[name] = {'best_score': current_score, 'best_params': clf_search.best_params_}

    print(f"Best {name} Classifier ROC-AUC (CV): {current_score:.4f}")
    print(f"Best {name} Classifier Params: {clf_search.best_params_}")

    if current_score > best_clf_score:
        best_clf_score = current_score
        best_clf_model = clf_search.best_estimator_
        best_clf_name = name

print(f"\n--- Best Classifier: {best_clf_name} with ROC-AUC (CV) of {best_clf_score:.4f} ---")
print("All Classifier Results:")
for name, res in clf_results.items():
    print(f"  {name}: ROC-AUC = {res['best_score']:.4f}")


best_reg_model = None
best_reg_score = np.inf # For RMSE (lower is better)
best_reg_name = ""
reg_results = {}

# Prepare regression target (only for purchases > 0)
X_train_reg = X_train[y_train_regress > 0].copy()
y_train_reg = y_train_regress[y_train_regress > 0].copy()
y_train_reg_log = np.log1p(y_train_reg)

print("\n--- Comparing Regression Models ---")
for name, config in regressors.items():
    print(f"\nTraining {name} Regressor...")

    # Conditional pipeline creation for SelectFromModel
    if name == 'XGBRegressor':
        # SelectFromModel with XGBoost for feature importance
        reg_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('selectfrommodel', SelectFromModel(estimator=XGBRegressor(random_state=42, n_jobs=-1, use_label_encoder=False))),
            ('regressor', config['model'])
        ])
    elif name in ['RandomForestRegressor', 'GradientBoostingRegressor']:
        # SelectFromModel with RandomForest for feature importance for other tree-based models
        reg_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('selectfrommodel', SelectFromModel(estimator=RandomForestRegressor(random_state=42, n_jobs=-1))),
            ('regressor', config['model'])
        ])
    else: # Fallback
        reg_pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('regressor', config['model'])
        ])

    reg_search = RandomizedSearchCV(
        reg_pipeline,
        config['params'], # 'config['params']' will contain appropriate 'selectfrommodel__threshold'
        n_iter=15,
        cv=KFold(n_splits=3, shuffle=True, random_state=42),
        scoring='neg_mean_squared_error',
        verbose=1,
        n_jobs=-1,
        random_state=42
    )
    reg_search.fit(X_train_reg, y_train_reg_log)

    current_rmse = np.sqrt(-reg_search.best_score_)
    reg_results[name] = {'best_rmse': current_rmse, 'best_params': reg_search.best_params_}

    print(f"Best {name} Regressor RMSE (log-transformed CV): {current_rmse:.4f}")
    print(f"Best {name} Regressor Params: {reg_search.best_params_}")

    if current_rmse < best_reg_score:
        best_reg_score = current_rmse
        best_reg_model = reg_search.best_estimator_
        best_reg_name = name

print(f"\n--- Best Regressor: {best_reg_name} with RMSE (log-transformed CV) of {best_reg_score:.4f} ---")
print("All Regressor Results:")
for name, res in reg_results.items():
    print(f"  {name}: RMSE = {res['best_rmse']:.4f}")

# --- Threshold tuning for classifier-regressor two-stage approach using best models ---
print("\n--- Performing Threshold Tuning for Best Combined Model ---")
val_is_purchase_probs = best_clf_model.predict_proba(X_val)[:, 1]
best_threshold = 0.5
best_val_r2 = -np.inf

threshold_range = np.concatenate([
    np.arange(0.01, 0.1, 0.01),
    np.arange(0.1, 0.9, 0.05),
    np.arange(0.9, 1.0, 0.01)
])

val_predictions = np.zeros(len(X_val))

for threshold in threshold_range:
    temp_val_predictions = np.zeros(len(X_val))
    val_purchase_indices = np.where(val_is_purchase_probs >= threshold)[0]
    if len(val_purchase_indices) > 0:
        X_val_for_reg = X_val.iloc[val_purchase_indices].copy()
        predicted_log_values = best_reg_model.predict(X_val_for_reg)
        predicted_actual_values = np.expm1(predicted_log_values)
        predicted_actual_values[predicted_actual_values < 0] = 0
        temp_val_predictions[val_purchase_indices] = predicted_actual_values
    temp_r2 = r2_score(y_val_regress, temp_val_predictions)
    if temp_r2 > best_val_r2:
        best_val_r2 = temp_r2
        best_threshold = threshold
        val_predictions = temp_val_predictions.copy()

print(f"\nBest Classifier Threshold on Validation Set: {best_threshold:.3f}")
print(f"Best Validation R2: {best_val_r2:.4f}")

# --- Final Prediction on Test Data using the Best Combined Model ---
print("\n--- Generating Final Predictions on Test Data ---")
test_is_purchase_probs = best_clf_model.predict_proba(X_test_final_submission)[:, 1]

final_test_predictions = np.zeros(len(X_test_final_submission))
purchase_indices = np.where(test_is_purchase_probs >= best_threshold)[0]

if len(purchase_indices) > 0:
    X_test_for_reg_final = X_test_final_submission.iloc[purchase_indices].copy()
    predicted_log_values = best_reg_model.predict(X_test_for_reg_final)
    predicted_actual_values = np.expm1(predicted_log_values)
    predicted_actual_values[predicted_actual_values < 0] = 0
    final_test_predictions[purchase_indices] = predicted_actual_values

print(f"Number of test sessions predicted to have a purchase: {len(purchase_indices)}")
if np.any(final_test_predictions > 0):
    print(f"Min predicted purchase value (non-zero): {np.min(final_test_predictions[final_test_predictions > 0]):.2f}")
else:
    print("No non-zero purchases predicted.")
print(f"Max predicted purchase value: {np.max(final_test_predictions):.2f}")

# --- Post-processing: Boost predicted values in 1e9 to 2.5e9 range by 1.2 (Validation & Test) ---
def boost_range(preds, lower=1e9, upper=2.5e9, factor=1.2):
    boosted = preds.copy()
    mask = (boosted >= lower) & (boosted < upper)
    boosted[mask] *= factor
    return boosted

# Apply boost to validation predictions
val_predictions = boost_range(val_predictions, lower=1e9, upper=2.5e9, factor=1.2)
# Apply boost to test predictions
final_test_predictions = boost_range(final_test_predictions, lower=1e9, upper=2.5e9, factor=1.2)

# --- Stacking: Fit a Ridge regression model only on nonzero predictions ---
# Validation set
val_nonzero_mask = val_predictions > 0
val_preds_base_nonzero = val_predictions[val_nonzero_mask].reshape(-1, 1)
y_val_true_nonzero = y_val_regress[val_nonzero_mask]

meta_model = Ridge()
if len(val_preds_base_nonzero) > 0:
    meta_model.fit(val_preds_base_nonzero, y_val_true_nonzero)

val_predictions_stacked = val_predictions.copy()
if len(val_preds_base_nonzero) > 0:
    val_predictions_stacked[val_nonzero_mask] = meta_model.predict(val_preds_base_nonzero)

# Test set
test_nonzero_mask = final_test_predictions > 0
test_preds_base_nonzero = final_test_predictions[test_nonzero_mask].reshape(-1, 1)
final_test_predictions_stacked = final_test_predictions.copy()
if len(test_preds_base_nonzero) > 0:
    final_test_predictions_stacked[test_nonzero_mask] = meta_model.predict(test_preds_base_nonzero)

# --- Diagnostics plot (validation, after stacking) ---
plt.figure(figsize=(10,5))
plt.hist(y_val_regress[y_val_regress>0], bins=100, alpha=0.5, label='True (Validation)', log=True)
plt.hist(val_predictions_stacked[val_predictions_stacked>0], bins=100, alpha=0.5, label='Stacked Predicted (Validation)', log=True)
plt.legend()
plt.xlabel('Purchase Value')
plt.ylabel('Frequency')
plt.title('Non-zero Purchase Value Distribution: True vs Stacked Predicted (Validation)')
plt.show()

# --- Plot: Predicted Purchase Value Distribution (Test set, after stacking) ---
plt.figure(figsize=(10,5))
plt.hist(final_test_predictions_stacked[final_test_predictions_stacked>0], bins=100, alpha=0.7, label='Stacked Predicted (Test)', log=True, color='orange')
plt.xlabel('Purchase Value')
plt.ylabel('Frequency')
plt.title('Predicted Non-zero Purchase Value Distribution (Test Set, Stacked)')
plt.legend()
plt.show()

# Evaluate new R2 on validation
r2_stacked = r2_score(y_val_regress, val_predictions_stacked)
print(f"\nValidation R2 after stacking/meta-model (nonzero only): {r2_stacked:.4f}")

# Create submission file (stacked predictions)
submission_df = pd.DataFrame({'id': range(len(final_test_predictions_stacked)), 'purchaseValue': final_test_predictions_stacked})
submission_df['purchaseValue'] = submission_df['purchaseValue'].round(2)
submission_filename = 'submission.csv'
submission_df.to_csv(submission_filename, index=False)

print(f"\nSubmission file '{submission_filename}' created successfully!")
print("First 5 rows of submission file:")
print(submission_df.head())
print("\nDescriptive statistics of purchaseValue in submission:")
print(submission_df['purchaseValue'].describe())
```
